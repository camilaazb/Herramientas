```python
# Variable "enfermedad"
dengue.enfermedad.value_counts()
```

```python
#cambiar la variable, para darle orden, gravedad
dengue['enfermedad_text']=dengue.enfermedad.astype(str)

dengue.replace({'enfermedad_text':{'SIN_SEÑALES':'1_SIN_SEÑALES','ALARMA':'2_ALARMA','GRAVE':'3_GRAVE'}},inplace=True)
```

```python
bd_2022 = dengue[dengue['ano'] == 2022]
```

Discretizing (agrupación, por edad)


```python
binLimits=[0,15,50,110]
theLabels=["a_menor_a_16","b_entre_16y50","c_mayor_a_50"]
bd_2022["edad_grupos"]=pd.cut(bd_2022['edad'], include_lowest=True,
                                     bins=binLimits,
                                     labels=theLabels,
                                     ordered=True)

# see

bd_2022.head()
```

```python
pd.crosstab( bd_2022.enfermedad_text,bd_2022.edad_grupos, dropna=False, normalize='columns')
```

```python
pd.crosstab(bd_2022.enfermedad_text,[bd_2022.sexo,bd_2022.edad_grupos], dropna=False, normalize='columns')
```

```python
df_distrit = df[df['UBIDIST'] == '100604'] #distro jose crespo, huanuco
```

```python
conteo_distrito = df_distrit['UBIDIST'].value_counts()

print(conteo_distrito)
```

```python
df_distrit['SEXO'] = df_distrit['SEXO'].replace({1: 'Femenino', 2: 'Masculino'})
```

```python
pd.crosstab( df_distrit.IDTIPOAC,df_distrit.SEXO, dropna=False, normalize='columns')
```

```python
df_distrit.ANO.value_counts()
```

```python
df_distrit['ANO'] = df_distrit['ANO'].astype(int) #convertir a numeros enteros
```

```python
df_distrit['IDTIPOAC'] = df_distrit['IDTIPOAC'].replace({'LDS': 'Desaparecido', 'TVS': 'Violacion', 'TLA': 'Lesion', 'LRC': 'Reclutamiento'})
```

```python
import pandas as pd
```

```python
print(df_distrit.columns)
```

Crear variable

```python
df_long = df_distrit.melt(id_vars=['ANO'], value_vars=['MAE', 'MAT', 'MEF', 'TVS', 'LDS_LDT', 'LRC', 'LSE'],
                          var_name='Afectacion', value_name='Ocurrencia')
```

```python
df_long = df_long[df_long['Ocurrencia'] == 1]
```

```python
df_grouped = df_long.groupby(['ANO', 'Afectacion']).size().reset_index(name='Cantidad de Víctimas')
```

Graficar

```python
# Crear un diccionario para mapear las etiquetas originales a las nuevas etiquetas
etiquetas_map = {
    'MAE': 'Ejecuciones extrajudiciales',
    'MAT': 'muertos en atentado',
    'MEF': 'muertos en enfrentamiento',
    'TVS': 'Violacion',
    'LSE': 'Secuestros',
    'LDS_LDT': 'Desaparecidos',
    'LRC': 'Reclutamiento'
    # Agrega más mapeos según sea necesario
}
```

```python
# Reemplazar los valores en la columna 'Afectacion'
df_grouped['Afectacion'] = df_grouped['Afectacion'].replace(etiquetas_map)
```

```python
grafico = alt.Chart(df_grouped).mark_line().encode(
    x='ANO:O',
    y='Cantidad de Víctimas:Q',
    color='Afectacion:N'
).properties(
    title='Cantidad de Víctimas por Año y Tipo de Afectación'
)

# Mostrar el gráfico
grafico
```

```python
indexList=['ANO','Afectacion']
aggregator={'Cantidad de Víctimas':['sum']}
by=df_grouped.groupby(indexList,observed=True).agg(aggregator)
by
```

```python
by_wide=by.unstack().fillna(0)
```

**Porcentaje (sgare)**
% de victimas desaparecidas

```python
sumCases=by_wide.sum(axis=1)
sumCases
```

```python
porcen=by_wide.loc[:,('Cantidad de Víctimas','sum','Desaparecidos')]/sumCases
porcen.name='Porcentaje'
porcen
```
